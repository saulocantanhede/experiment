```python
from tf.app import use
```

```python
BHS = use("etcbc/bhsa", version=2021, hoist=globals())
```

### August 28, 2024 - Genesis 22:1-6

```python
results = BHS.search("""
book book=Genesis
    chapter chapter=22
        verse gloss*
""")
BHS.show(results, end=6, multiFeatures=False, queryFeatures=True, condensed=True)
```

```python
results = BHS.search("""
word gloss* g_cons=NCTXWH vs* vt* nu* ps*
""")
BHS.show(results, end=1, multiFeatures=False, queryFeatures=True, condensed=True)
```

```python
results = BHS.search("""
word gloss* g_cons=NSH
""")
BHS.show(results, end=4, multiFeatures=False, queryFeatures=True, condensed=True)
```

### September 4, 2024 - Genesis 22:7-16

Action of "lifted up eyes and look"

```python
results = BHS.search("""
book book=Genesis
    chapter chapter*
        verse verse*
            w1:word lex=NF>[ gloss*
            w2:word lex=R>H[
w2 > w1
""")
BHS.table(results, multiFeatures=False, queryFeatures=True, condensed=True)
```

```python
results = BHS.search("""
book book*
    chapter chapter*
        verse verse*
            sentence
                w1:word lex=R>H[ gloss* vs=qal
                w2:word lex=L prs*
w1 <: w2
""")
BHS.show(results, multiFeatures=False, queryFeatures=True, condensed=True)
```
