```python
nodes_file_path = '../new_data/phrase'

phrase_boundaries = []

with open(nodes_file_path, 'r') as f:
    for line in f:
        line = line.strip()
        if '-' in line:
            start, end = map(int, line.split('-'))
            phrase_boundaries.append((start, end))
        else:
            phrase_boundaries.append((int(line), int(line)))
```

```python
phrase_boundary = {}

def find_phrase_boundary(node):
    for start, end in phrase_boundaries:
        if start <= node <= end:
            return f"{start}-{end}"

for node in F.otype.s('word'):                    
    boundary = find_phrase_boundary(node)
    if boundary:
        phrase_boundary[node] = boundary
```
