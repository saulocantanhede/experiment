#### Broken `frame` references

```python
id_map = {}

for node, item in NT.api.F.id.items():
    id_map[item] = node

valid_nodes = set(id_map.values())

def parse_framespec_string(spec):
    if not spec or spec == '{}':
        return {}
    s = spec.strip()
    if s.startswith('{') and s.endswith('}'):
        s = s[1:-1]
    out = {}
    for part in re.split(r'\s*,\s*', s):
        if not part:
            continue
        role, ids = part.split(':', 1)
        role = role.strip()
        for id_str in re.split(r'\s*;\s*', ids.strip()):
            if not id_str:
                continue
            out[int(id_str)] = role
    return out

framedict = {}
bad_raw_ids = {}        # collect unmapped raw ids for debugging
for frame_node, spec in NT.api.F.framespec.items():
    parsed = parse_framespec_string(spec)
    if not parsed:
        continue

    subdict = {}
    for raw_id, role in parsed.items():
        # interpret raw_id == 0 as pointing to the frame node itself
        if raw_id == 0:
            # try to map the frame node to a dataset node (if frame_node is an "id" value)
            # often key is already the dataset node number; if not, try int(frame_node)
            candidate = int(frame_node)
            mapped = id_map.get(candidate, candidate)  # try map; fallback to candidate
        else:
            mapped = id_map.get(raw_id)  # only map if raw_id is in id_map

        # Accept mapped only if it's a valid dataset node
        if mapped in valid_nodes:
            subdict[mapped] = role
        else:
            bad_raw_ids[frame_node] = raw_id

    if subdict:
        framedict[frame_node] = subdict

if bad_raw_ids:
    print(f"There are {len(bad_raw_ids)} broken frame references.")
    for bad_raw_id in bad_raw_ids:
        #print(bad_raw_id)
        ref = NT.api.T.sectionFromNode(bad_raw_id)
        print(f"{ref[0]} {ref[1]}:{ref[2]} - {bad_raw_ids[bad_raw_id]}")
```

```python
import re
import requests
from bs4 import BeautifulSoup, Tag, NavigableString
from typing import Dict, Tuple, Optional

VERSE_PATTERNS = [
    re.compile(r'_\{\s*(\d+)\s*\}'),     # _{18}
    re.compile(r'\{\s*(\d+)\s*\}'),      # {18}
    re.compile(r'^\s*(\d+)\s*$'),        # standalone number
    re.compile(r'v(?:erse)?\.?\s*(\d+)', re.I),  # "v. 18" or "verse 18"
]

# also match id/name patterns like "...MAT.1.18" or "...MAT_1_18" or "V18"
ID_PAT = re.compile(r'(\d+)(?:$|[^\d])')

def find_first_verse_after(tag, max_tokens=200) -> Optional[int]:
    """
    Scan forward from `tag` among next_elements (up to max_tokens items)
    trying to detect the first verse number using several heuristics.
    Returns int verse or None.
    """
    seen = 0
    for elem in tag.next_elements:
        if seen >= max_tokens:
            break
        seen += 1

        # direct text checks
        if isinstance(elem, NavigableString):
            text = elem.strip()
            if text:
                for pat in VERSE_PATTERNS:
                    m = pat.search(text)
                    if m:
                        try:
                            return int(m.group(1))
                        except ValueError:
                            pass

                # also try any standalone integer in the text
                m2 = re.search(r'\b(\d{1,3})\b', text)
                if m2:
                    return int(m2.group(1))

        # element attribute checks (id / name / href anchors)
        if isinstance(elem, Tag):
            # check common attributes
            for attr in ("id", "name", "href", "data-anchor"):
                val = elem.get(attr)
                if val:
                    # find trailing number in id/name/href (e.g. MAT.1.18 or #18)
                    m = re.search(r'(\d+)(?:$|[^\d])', val)
                    if m:
                        try:
                            return int(m.group(1))
                        except ValueError:
                            pass

            # check text inside special small/sup/span tags quickly
            inner_text = elem.get_text(strip=True)
            if inner_text:
                for pat in VERSE_PATTERNS:
                    m = pat.search(inner_text)
                    if m:
                        try:
                            return int(m.group(1))
                        except ValueError:
                            pass

    return None


def sections_to_dict(url: str, heading_tags=("h4",), verbose: bool = False) -> Dict[Tuple[str,int,int], str]:
    """
    Fetch page at `url`, extract headings and the first verse of the section.
    Returns dict keyed by (BOOK, chapter, verse) -> title
    If the verse cannot be found, uses an incremental fallback starting at 1.
    """
    # infer book and chapter from URL like .../UBS5/MAT.1
    m = re.search(r'/([A-Z0-9]+)\.(\d+)$', url.rstrip('/'))
    if not m:
        print(m, url)
        raise ValueError("URL doesn't match expected pattern '/XXX.N' (e.g. /MAT.1)")
    book, chapter = m.group(1), int(m.group(2))

    html = requests.get(url, timeout=30).text
    soup = BeautifulSoup(html, "lxml")

    headings = []
    for tagname in heading_tags:
        headings.extend(soup.find_all(tagname))

    results: Dict[Tuple[str,int,int], str] = {}
    fallback_index = 1

    for h in headings:
        title = h.get_text(" ", strip=True)
        # skip empty or obviously parenthetical headings
        if not title:
            continue
        if title.startswith("(") and title.endswith(")"):
            continue

        verse = find_first_verse_after(h)
        if verse is None:
            # fallback: try to find nearest preceding verse number instead
            verse = None
            # search backward a small number of tokens
            seen = 0
            for elem in h.previous_elements:
                if seen >= 80:
                    break
                seen += 1
                if isinstance(elem, NavigableString):
                    text = elem.strip()
                    if text:
                        m2 = re.search(r'\b(\d{1,3})\b', text)
                        if m2:
                            try:
                                verse = int(m2.group(1))
                                break
                            except ValueError:
                                pass
                if isinstance(elem, Tag):
                    txt = elem.get_text(strip=True)
                    if txt:
                        m2 = re.search(r'\b(\d{1,3})\b', txt)
                        if m2:
                            try:
                                verse = int(m2.group(1))
                                break
                            except ValueError:
                                pass

        if verse is None:
            # final fallback: incremental index so keys remain unique
            verse = fallback_index
            fallback_index += 1

        key = (book, chapter, int(verse))
        # avoid overwriting an earlier heading with identical verse: append suffix if needed
        if key in results:
            # choose next integer verse (simple collision resolution)
            kbook, kchap, kverse = key
            newv = kverse + 1
            while (kbook, kchap, newv) in results:
                newv += 1
            key = (kbook, kchap, newv)

        results[key] = title
        if verbose:
            print(f"Found: {key} -> {title}")

    return results


# Example usage with NT.api iteration (preserves your original approach).
# NOTE: NT.api must be available in the runtime where you run this.
def collect_all_sections(NT_api) -> Dict[Tuple[str,int,int], str]:
    """
    Given NT.api (as in your code), iterate books and chapters building a single dict.
    Returns a dict mapping (book, chapter, verse) -> title
    """
    all_sections = {}
    names = []
    # you had for book, item in NT.api.F.book.items():
    # assuming the API matches that pattern:
    for book, item in NT_api.F.book.items():
        # iterate chapters using your API
        for chapter in NT_api.F.L.d(book, 'chapter'):    # keep your original API call
            num = NT_api.F.chapter.v(chapter)
            url = f'https://www.die-bibel.de/en/bible/UBS5/{item}.{num}'
            try:
                page_sections = sections_to_dict(url)
            except Exception as e:
                # continue but record error info in verbose runs
                print(f"Error fetching {url}: {e}")
                continue

            # merge into master dict (beware of collisions across books/chapters)
            for k, v in page_sections.items():
                if k in all_sections:
                    # unlikely if k contains book and chapter, but guard anyway
                    bk, ch, vrs = k
                    newv = vrs + 1
                    while (bk, ch, newv) in all_sections:
                        newv += 1
                    all_sections[(bk, ch, newv)] = v
                else:
                    all_sections[k] = v

            names.append(page_sections)   # keeps per-page returned dicts if you want them

    return all_sections

# If you want a flat list of titles instead of keyed dict, you can do:
#flat_titles = list(all_sections.values())
```

```python
import re
import requests
from bs4 import BeautifulSoup

def sections_to_dict(url: str):
    """
    Extrai os títulos das seções e os mapeia para o número do primeiro
    versículo daquela seção, usando o método find_next() para garantir a ordem.
    """
    m = re.search(r'/([A-Z0-9]+)\.(\d+)$', url.rstrip('/'))
    if not m:
        return {}
    book, chapter = m.group(1), int(m.group(2))

    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, timeout=30, headers=headers)
        response.raise_for_status()
        html = response.text
        soup = BeautifulSoup(html, "lxml")
    except requests.RequestException as e:
        print(f"Erro ao buscar a URL: {e}")
        return {}

    headings = []
    for h in soup.find_all("h4"):
        title = h.get_text(strip=True)
        if not title or (title.startswith("(") and title.endswith(")")):
            continue
        headings.append(h)

    results = {}

    # --- INÍCIO DA CORREÇÃO FINAL ---

    # Regex para encontrar e extrair o número do versículo de um ID. Ex: id="UBS5_MAT.1.18"
    # O padrão foi ajustado para corresponder ao ID completo, como 'UBS5_MAT.1.1', para mais precisão.
    verse_id_pattern = re.compile(rf'.*?{book}\.{chapter}\.(\d+)')

    for h in headings:
        title = h.get_text(strip=True)
        
        # A partir do elemento do título 'h', encontre a próxima tag
        # (de qualquer tipo) cujo atributo 'id' corresponda ao nosso padrão de versículo.
        next_verse_element = h.find_next(id=verse_id_pattern)

        if next_verse_element:
            # Se um elemento foi encontrado, aplicamos o regex no seu ID para extrair o número.
            match = verse_id_pattern.search(next_verse_element.get('id', ''))
            if match:
                verse_number = int(match.group(1))
                
                # Uma verificação extra para evitar adicionar o mesmo título com versículos diferentes
                # (caso haja múltiplos versículos antes do próximo título)
                if title not in results.values():
                    results[verse_number] = title
    
    # --- FIM DA CORREÇÃO FINAL ---

    return results

# Exemplo com a sua URL:
url = "https://www.die-bibel.de/en/bible/UBS5/MAT.1"
bible_sections = sections_to_dict(url)
print(bible_sections)
```

We saved the `ref` feature as `referentspec`, `refenglish` as `referent_translation_english`, `refindex` as `referent_index`, `reflemma` as `referent_lemma`, and `refstrong` as `referent_strong`.\
Also, we save `ref` values as an edge feature.

```python
referentspec = {}
referent_translation_english = {}
referent_index = {}
referent_lemma = {}
referent_strong = {}

for node, item in NT.api.F.ref.items():
    if item != '{}' and item:
        referentspec[node] = item.replace('{','').replace('}', '').replace(' ', ', ')

for node, item in NT.api.F.refenglish.items():
    if item:
        item = item.replace('{','').replace('}', '')
        if '~' in item:
            referent_translation_english[node] = re.sub(r'\s*~\s*', ' ', item).strip("~ ").strip()
            if '*' in referent_translation_english[node]:
                referent_translation_english[node] = re.sub(r'\s*\*\s*', ' ', referent_translation_english[node]).strip("* ").strip()
        elif '*' in item:
            referent_translation_english[node] = re.sub(r'\s*\*\s*', ' ', item).strip("* ").strip()
            if '-' in referent_translation_english[node]:
                referent_translation_english[node] = re.sub(r'\s*~\s*', ' ', referent_translation_english[node]).strip("* ").strip()
        else:
            referent_translation_english[node] = item

for node, item in NT.api.F.refindex.items():
    if item:
        referent_index[node] = re.sub(r'[^0-9]', '', item)

for node, item in NT.api.F.reflemma.items():
    if item != '{}' and item:
        referent_lemma[node] = item.replace('{','').replace('}', '').replace(' ', ', ')

for node, item in NT.api.F.refstrong.items():
    if item != '{}' and item:
        referent_strong[node] = item.replace('{','').replace('}', '').replace(' ', ', ')
```

```python
id = {}

for node, item in NT.api.F.id.items():
    id[item] = node

referentdict = {}

for wordnode in NT.api.F.otype.s('word'):
    item = NT.api.F.ref.v(wordnode)
    if item != '{}' and item:
        referentnode = []
        refentid = item.replace('{','').replace('}', '').split(' ')
        for node in refentid:
            if node[-1] == ',':
                node = node[:-1]
            referentnode.append(id[int(node)])
        referentnode = str(referentnode).replace('[','{').replace(']','}')
        referentdict[wordnode] = referentnode
```

We saved the `subjref` values in a feature called `subjrefspec`, `subjrefenglish` as `subjref_translation_english`, `subjreflemma` as `subjref_lemma`, and `subjrefstrong` as `subjref_strong`.\
Also, we save `subjref` as an edge feature.

```python
subjrefspec = {}
subjref_translation_english = {}
subjref_lemma = {}
subjref_strong = {}

for node, item in NT.api.F.subjref.items():
    if item != '{}' and item:
        subjrefspec[node] = item.replace('{','').replace('}', '').replace(' ', ', ')

for node, item in NT.api.F.subjrefenglish.items():
    if item:
        item = item.replace('{','').replace('}', '')
        if '~' in item:
            subjref_translation_english[node] = re.sub(r'\s*~\s*', ' ', item).strip("~ ").strip()
            if '*' in subjref_translation_english[node]:
                subjref_translation_english[node] = re.sub(r'\s*\*\s*', ' ', subjref_translation_english[node]).strip("* ").strip()
        elif '*' in item:
            subjref_translation_english[node] = re.sub(r'\s*\*\s*', ' ', item).strip("* ").strip()
            if '-' in subjref_translation_english[node]:
                subjref_translation_english[node] = re.sub(r'\s*~\s*', ' ', subjref_translation_english[node]).strip("* ").strip()
        else:
            subjref_translation_english[node] = item

for node, item in NT.api.F.subjreflemma.items():
    if item != '{}' and item:
        subjref_lemma[node] = item.replace('{','').replace('}', '').replace(' ', ', ')

for node, item in NT.api.F.subjrefstrong.items():
    if item != '{}' and item:
        subjref_strong[node] = item.replace('{','').replace('}', '').replace(' ', ', ')
```

```python
id = {}

for node, item in NT.api.F.id.items():
    id[item] = node

subjrefdict = {}

for wordnode in NT.api.F.otype.s('word'):
    item = NT.api.F.subjrefspec.v(wordnode)
    if item != '{}' and item:
        nums = re.findall(r'\d+', item)
        subjrefset = { id[int(n)] for n in nums }
        subjrefdict[wordnode] = subjrefset
```
